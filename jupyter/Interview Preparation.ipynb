# Arrays

Python arrays are ordered collections that store multiple elements of the same data type. Unlike Python lists which can store various data types simultaneously, <mark>arrays are more memory-efficient</mark> but require importing the array module.

**Key Characteristics:**
- Arrays must contain elements of identical data type
- Arrays are mutable and can be modified after creation
- Arrays require importing the array module, as they are not built-in like lists

Below is a typecode table, with the different typecodes that can be used with the different data types when defining Python arrays:




```python
import array

int_array = array.array('i', [53, 54, 55])
print(int_array)
```
